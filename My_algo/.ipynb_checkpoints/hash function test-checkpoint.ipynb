## kosarak

```python
# spooky hash
# input : binary, encoding is parameter of bytes()
# output: unsigned- 32 bit int

# distinct: 606770
# hashed : 606715
# collision rate: 0.0090644% 
# Execution time:0.7328388690948486 seconds.


import spookyhash
import pandas as pd
import os

filename='kosarak_ground_truth.csv'
filepath="..\\result\\kosarak\\"
src_data=os.path.join(filepath,filename)
distinct=606770

df=pd.read_csv(src_data)
inputli=list(df['Element'])
itemset=set()
itemdict=dict()

start=time.time()
# use hashed value as key
for item in inputli:
    x=spookyhash.hash32(bytes((item),encoding='utf-8'))
    if x in itemdict:
        print("{} collision with {}.".format(itemdict[x],item))
    else:
        itemdict[x]=item
end=time.time()
hashed_item=len(itemdict)
print("# of hashed item: {}".format(hashed_item))
print("collision rate: {:.7f}% ".format((distinct-hashed_item)*100/distinct))
    # 以非科學表示法output小數
print("collision rate: {}".format((distinct-hashed_item)/distinct))
print("Execution time:{} seconds.".format(str(end-start)))
```

```python
# murmur hash
# input : str
# output: unsigned- 32 bit int

# distinct: 606770
# hashed item: 606736
# collision rate: 0.0056034%
# Execution time:0.6411628723144531 seconds.

import pandas as pd
import os
import mmh3
import time

filename='kosarak_ground_truth.csv'
filepath="..\\result\\kosarak\\"
src_data=os.path.join(filepath,filename)
distinct=606770



df=pd.read_csv(src_data)
inputli=list(df['Element'])
itemdict=dict()

start=time.time()
# use hashed value as key
for item in inputli:
    x=mmh3.hash(item, signed=False)
    if x in itemdict:
        print("{} collision with {}.".format(itemdict[x],item))
    else:
        itemdict[x]=item
end=time.time()
hashed_item=len(itemdict)
print("# of hashed item: {}".format(hashed_item))
print("collision rate: {:.7f}% ".format((distinct-hashed_item)*100/distinct))
    # 以非科學表示法output小數
print("collision rate: {}".format((distinct-hashed_item)/distinct))
print("Execution time:{} seconds.".format(str(end-start)))
```
