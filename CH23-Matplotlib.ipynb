```python
#基本流程： 使用plt.figure定义一个图像窗口. 使用plt.plot画(x ,y)曲线. 使用plt.show显示图像.

#把list傳給pyplot.plot(), list內的data為Y軸的值，X軸則由list的index產生
import matplotlib.pyplot as plt

square=[1,4,9,16,25,36,49,64,81]
plt.plot(square)
plt.show()
```

```python
#axis()更改x和y軸的最小與最大刻度

import matplotlib.pyplot as plt

square=[0,1,4,9,16,25,36,49,64,81]
plt.plot(square)
plt.axis([0,8,0,70]) #([x-min,x-max,y-min,y-max])
plt.show()
```

```python
#線條寬度linewidth or lw
#matplotlib 不支援中文顯示 
#plt.title()  參數：https://blog.csdn.net/helunqu2017/article/details/78659490
#plt.xlabel("")
#plt.ylabel()
import matplotlib.pyplot as plt

square=[0,1,4,9,16,25,36,49,64,81]
plt.plot(square,lw=5)    #plot()一堆參數:https://blog.csdn.net/claroja/article/details/70792880
plt.axis([0,8,0,70])

plt.title("Square Chart",fontsize=24)
plt.xlabel("Value")
plt.ylabel("Square")

plt.show()
```
