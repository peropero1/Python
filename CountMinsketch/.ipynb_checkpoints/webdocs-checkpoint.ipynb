```python
# count distinct element
num_line=0
count=0
itemdict={}
filename='webdocs.dat'
filepath=r"X:\\NTU\\ML-sketch\\dataset\\Webpage\\webdocs\\"


with open(filepath+filename,'r') as file:
    while True:
        line=file.readline().strip('\n')
        num_line+=1
        if not line:
            break
        else:
            if line not in itemdict:
                itemdict[line]=1
            else:
                itemdict[line]+=1
print("Distinct element: {}".format(len(itemdict)))
print("Number of incomming element: {}".format(num_line))

with open("webdocs_distinct_count.txt",'w') as write_obj:
    for item in sorted(itemdict.items(),key = lambda s: s[1],reverse=True):
        temp=list(item)
        tempstr=temp[0]+','+str(temp[1])+'\n'
        write_obj.write(tempstr)
```

## Space-Saving
- 10000 items:
    - Top-1024 with size 2416 bytes.
    - Execution time:0.8078114986419678 seconds.
- Full test
    - Top-1024 with size 4720 bytes.
    - Execution time:305.27067971229553 seconds.

```python
# space saving with webdocs, top-1024 elements,read 10000 items.

import sys
import time


start=time.time()
size=1024
itemlist=[]
item=10000

filename='webdocs.dat'
filepath=r"X:\\NTU\\ML-sketch\\dataset\\Webpage\\"

with open(filepath+filename,'r') as file:
    while item:
        line=file.readline().strip('\n')
        if not line:
            break
        else:
            item-=1
            if len(itemlist)==0:
                itemlist.append([line,1])
            else:
                for i in range(len(itemlist)):
                    if line==itemlist[i][0]:
                        itemlist[i][1]+=1
                        found=True
                        #print("found in itemlist[{}]".format(i))
                        break
                    else:
                        found=False
                if found:
                    pass
                else:
                    if len(itemlist)<size:
                        itemlist.append([line,1])
                    else:
                        # len(itemset)=size
                        last=len(itemlist)-1
                        itemlist[last][0]=line
                        itemlist[last][1]+=1
            itemlist=sorted(itemlist,key = lambda itemlist: itemlist[1],reverse=True)
        #print("itemlist:{}\nlen(itemlist):{}\n".format(itemlist,len(itemlist)))
end=time.time()
print(itemlist[:64],len(itemlist))
print("Top-{} with size {} bytes.".format(len(itemlist),sys.getsizeof(itemlist)))
print("Execution time:{} seconds.".format(str(end-start)))
```

```python
# space saving with webdocs, top-1024 elements, full read

import sys
import time


start=time.time()
size=1024
itemlist=[]

filename='webdocs.dat'
filepath=r"X:\\NTU\\ML-sketch\\dataset\\Webpage\\"

with open(filepath+filename,'r') as file:
    while True:
        line=file.readline().strip('\n')
        if not line:
            break
        else:
            if len(itemlist)==0:
                itemlist.append([line,1])
            else:
                for i in range(len(itemlist)):
                    if line==itemlist[i][0]:
                        itemlist[i][1]+=1
                        found=True
                        #print("found in itemlist[{}]".format(i))
                        break
                    else:
                        found=False
                if found:
                    pass
                else:
                    if len(itemlist)<size:
                        itemlist.append([line,1])
                    else:
                        last=len(itemlist)-1
                        itemlist[last][0]=line
                        itemlist[last][1]+=1
            itemlist=sorted(itemlist,key = lambda itemlist: itemlist[1],reverse=True)
        #print("itemlist:{}\nlen(itemlist):{}\n".format(itemlist,len(itemlist)))
end=time.time()
print(itemlist[:64],len(itemlist))
print("Top-{} with size {} bytes.".format(len(itemlist),sys.getsizeof(itemlist)))
print("Execution time:{} seconds.".format(str(end-start)))
```

```python
# distinct element count

filename='webdocs.dat'
filepath=r"X:\\NTU\\ML-sketch\\dataset\\Webpage\\"
count=0
itemset=set()

with open(filepath+filename,'r') as file_obj:
    while True:
        line=file_obj.readline().strip('\n')
        if not line:
            break
        else:
            if line not in itemset:
                itemset.add(line)
                count+=1
print("distinct element: {}".format(count))

with open("webdocs.txt",'w') as write_obj:
    for i in itemset:
        write_obj.write(i)
```
