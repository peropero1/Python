```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('classic')
    # 設定樣式
fig=plt.figure()
    # 
x=np.linspace(0,10,100)
    # start,stop,num
print(x)
    # x為一串列

    
plt.plot(x,np.sin(x),'-')
    # 畫第一條線，為把串列做為參數傳給plot, 串列內的資料會被視為y軸的值，x軸的值會依串列值的index產生
plt.plot(x,np.cos(x),'--')
    # 第二條
plt.show()

fig.savefig("sin_cos.png")
    # 儲存圖檔到當前的工作目錄
```

```python
# 多組數據 plt(index1,seq1,index2,seq2)
# https://ithelp.ithome.com.tw/articles/10232059

import matplotlib.pyplot as plt
import os
import pandas as pd
import numpy

size=1024

filepath=r"..\dataset\kosarak"
gr_file_name='kosarak_ground_truth.csv'
gr_path=os.path.join(filepath,gr_file_name)

result_dir=r"..\result"

my_file='My_kosarak_Top_1024_8_512.csv'
my_path=os.path.join(result_dir,my_file)

ela_file='Ela_kosarak_Top-1024_8_1024.csv'
ela_path=os.path.join(result_dir,ela_file)


gr_df=pd.read_csv(gr_path)
my_df=pd.read_csv(my_path)
ela_df=pd.read_csv(ela_path)

gr_list=gr_df['Count'][:size].values.tolist()
my_list=my_df['Count'][:size].values.tolist()
ela_list=ela_df['Count'][:size].values.tolist()
ela_list.sort(reverse=True)

gr_list=np.log(gr_list)
my_list=np.log(my_list)
ela_list=np.log(ela_list)


# 設定figure
indexli=[i for i in range (1024)]
plt.xlabel('Index')
plt.ylabel('Count, log scale')
plt.plot(indexli,gr_list,'r',indexli,my_list,'g',indexli,ela_list,'b')
    # 多組資料為index1,data1,index2,data2,...
plt.show()

```

```python
# legend
# 加入說明各顏色代表哪個algo
# 設定每個刻度的間隔為50

import matplotlib.pyplot as plt
import os
import pandas as pd
import numpy

size=1024

filepath=r"..\dataset\kosarak"
gr_file_name='kosarak_ground_truth.csv'
gr_path=os.path.join(filepath,gr_file_name)

result_dir=r"..\result"
my_file='My_kosarak_Top_1024_8_512.csv'
my_path=os.path.join(result_dir,my_file)

ela_file='Ela_kosarak_Top-1024_8_1024.csv'
ela_path=os.path.join(result_dir,ela_file)


gr_df=pd.read_csv(gr_path)
my_df=pd.read_csv(my_path)
ela_df=pd.read_csv(ela_path)

gr_list=gr_df['Count'][:size].values.tolist()
my_list=my_df['Count'][:size].values.tolist()
ela_list=ela_df['Count'][:size].values.tolist()
ela_list.sort(reverse=True)

gr_list=np.log(gr_list)
my_list=np.log(my_list)
ela_list=np.log(ela_list)
indexli=[i for i in range (1024)]
plt.xticks([j for j in range(0,1200,100)])
plt.xlabel('Index')
plt.ylabel('Count, log scale')

gr_line,=plt.plot(indexli,gr_list,'r',label='GroundTruth')
my_line,=plt.plot(indexli,my_list,'g',label='My algo')
ela_line,=plt.plot(indexli,ela_list,'b',label='Elastic Sketch')
    # 注意此處的,= 且必須同時設定label
plt.legend(handles=[gr_line,my_line,ela_line],loc='best')
    # 設定子圖


plt.savefig('trial.png')    
plt.show()
```
