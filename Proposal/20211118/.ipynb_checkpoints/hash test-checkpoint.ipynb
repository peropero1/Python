```python
import sys
sys.path.append("..")
import os
import mmh3
from Crypto.Util import number
import numpy as np

seed=[number.getPrime(32) for i in range(2)]

filename='kosarak.dat'
filepath=r"..\dataset\kosarak"
src_data=os.path.join(filepath,filename)
    # dataset

w=1024
d=8
r=w*d

sketch_2d=np.zeros((d,w),dtype='int32')
sketch_1d=np.zeros(r,dtype='int32')

item_set=set()
collision_1d=0
collision_2d=0

item_count=100000
with open(src_data,'r') as file:
    while item_count:
        e=file.readline().strip('\n')
        if not e:
            print('EOF')
            break
        else:
            item_count-=1
            hashv=mmh3.hash(e,seed=seed[0], signed=False)% r
            row=mmh3.hash(e,seed=seed[0], signed=False)% d
            col=mmh3.hash(e,seed=seed[1], signed=False)% w
            if e not in item_set:
                if sketch_1d[hashv]==1:
                    collision_1d+=1
                else:
                    sketch_1d[hashv]=1
                if sketch_2d[row][col]==1:
                    collision_2d+=1   
                else:
                    sketch_2d[row][col]=1  
                item_set.add(e)
print(collision_1d,collision_2d)
```

```python
import sys
sys.path.append("..")
import os
import mmh3
from Crypto.Util import number
import numpy as np
import re

seed=[number.getPrime(32) for i in range(2)]

filepath=r"..\dataset\caida"
pattern='\w+.dat$'
r=re.compile(pattern)
filelist=list(filter(r.match,os.listdir(filepath)))  
    # dataset

w=1024
d=8
r=w*d

sketch_2d=np.zeros((d,w),dtype='int32')
sketch_1d=np.zeros(r,dtype='int32')

item_set=set()
collision_1d=0
collision_2d=0
item_count=100000

for datafile in filelist[:1]:
    src_data=os.path.join(filepath,datafile)
    with open(src_data,'rb') as file:
         #以binary讀取，資料型態也為byte
        while item_count:
            e=str(file.read(13))
            if len(e)<13:
                print('EOF')
                break
            else:
                item_count-=1
                hashv=mmh3.hash(e,seed=seed[0], signed=False)% r
                row=mmh3.hash(e,seed=seed[0], signed=False)% d
                col=mmh3.hash(e,seed=seed[1], signed=False)% w
                if e not in item_set:
                    if sketch_1d[hashv]==1:
                        collision_1d+=1
                    else:
                        sketch_1d[hashv]=1
                    if sketch_2d[row][col]==1:
                        collision_2d+=1   
                    else:
                        sketch_2d[row][col]=1  
                    item_set.add(e)
print(collision_1d,collision_2d)
```
