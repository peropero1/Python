```python
import heapq

import sys
sys.path.append("..")

import Tools.Config as TC
import Tools.Func as TF
import Node.DS as DS
import time
import os

from pympler import asizeof
import re
import pandas as pd
dataset='kosarak'
total_count=0
heavy_ratio=1/5000

def index_of_element(item_list,item):
    for i,element in enumerate(item_list):
        if element[1]==item:
            return i
    return None

# =============================dataset path and file=============================
filename=dataset+'.dat'
filepath=r"..\..\dataset\kosarak"
gr_file_name='kosarak_ground_truth.csv'
src_data=os.path.join(filepath,filename)
    # dataset
gr_path=os.path.join(filepath,gr_file_name)
    # ground truth
    
# =============================Initialize=============================
w=0
d=0
size=5000
topk=5000
TC.Set_default(w,d,size,topk)
    # set width, depth, size of Sk, random seed of hash
    # Config.width, Config.depth
Top_dict=dict()
    # Top_dict[x]=[count_x,error_x]
ss_heap=[]
# ss_heap[i]=[count,id,error]
item_count=50000

# =============================Stream processing=============================
start=time.time()
with open(src_data,'r') as file:
    while True:
        e=file.readline().strip('\n')
        #item_count-=1
        if not e:
            print('EOF')
            break
        else:
            total_count+=1
            find=index_of_element(ss_heap,e)
            if find!=None:
                # e is in ss_heap
                ss_heap[find][0]+=1
                heapq.heapify(ss_heap)
            else:
                if len(ss_heap)<TC.size:
                    # ss_heap is not full
                    heapq.heappush(ss_heap,[1,e,0])
                else:
                    # ss_heap is full
                    min_count,min_ele,min_err=heapq.heappop(ss_heap)
                    # [count,id]
                    heapq.heappush(ss_heap,[min_count+1,e,min_count])
end=time.time()
print("Execution time:{:8.3f} seconds.".format(end-start))
print("Memory usage: {:.3f} kbytes.".format(asizeof.asizeof(ss_heap)/1024))

# =============================heavy hitter from result=============================
# heavy hitter candidates
HH=dict()
for item in ss_heap:
    if item[0]>total_count*heavy_ratio:
        if (item[0]-item[2])>=total_count*heavy_ratio:
            HH[item[1]]=item[0]
HH=dict(sorted(HH.items(), key=lambda item: item[1],reverse=True))

# ground truth
import pandas as pd
df=pd.read_csv(gr_path)
temp=df[df['Count']>=int(total_count*heavy_ratio)]

gr_set=set(temp['Element'])
result=set(HH.keys())
tp_set=gr_set & result

# jaccard similarity
jaccard=len(tp_set)/(len(gr_set)+len(result)-len(tp_set))

# Count ARE/AAE in Top
startx=time.time()
top_are,top_aae=TF.Get_ARE_AAE(gr_path,HH,tp_set)
print("Find:{}".format(len(tp_set)))
#print("{} item found in SS[{}] compare with true Top-{}".format(len(tp_set),Config.size,topk))
endx=time.time()

print("\nFor copy")
print("Execution time:{:8.3f} seconds.".format(end-start))
print("Top-{}".format(TC.size))
print("Top_dict with {:.3f} kbytes.".format(asizeof.asizeof(ss_heap)/1024))
print("Heavy Hitter Candidates: {}".format(len(HH)))
print("Heavy Hitter Ground Truth: {}".format(len(temp)))
print("Jaccard similarity:{}".format(jaccard))
print("HH_ARE: {:6.4f}".format(top_are))
print("HH_AAE: {:6.4f}".format(top_aae))

print("\n{:<7}{:<10}{:<13}{:<13}{:<10}{:<10}".format('Top','Memory','Precision','Execution','HH_ARE','HH_AAE'))
print("{:<7}{:<10.3f}{:<13.4f}{:<13.3f}{:<10.4f}{:<10.4f}".format(TC.size,asizeof.asizeof(Top_dict)/1024,jaccard,end-start,top_are,top_aae))

TF.Plot_hh_compare(temp,HH,"SpaceSaving")
```
