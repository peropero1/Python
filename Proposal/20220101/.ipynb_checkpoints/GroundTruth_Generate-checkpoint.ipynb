# Caida 
## 1030031 distinct flows

## 27553679 arrivals

```python
# Generate caida_xx_ground_truth.csv
import sys
sys.path.append("..")

import Tools.Config as TC
import Tools.Func as TF
import Node.DS as DS
import numpy as np
import time
import os
import hyperloglog

import pandas as pd
import re

dataset='caida'
filepath=r"..\dataset\caida"
pattern='\w+.dat$'
r=re.compile(pattern)
filelist=list(filter(r.match,os.listdir(filepath)))    
    # dataset file list

# 產生groudtruth的檔名 caida_XX_ground_truth.csv
filename=[]
for item in filelist:
    fid=re.search('\d\d', item).group(0)
    filename.append('caida_'+fid+'_ground_truth.csv')    

#filename='caida'+'_ground_truth.csv'
item_dict=dict()
item_set=set()
#item_count=20
for datafile,fileindex in zip(filelist[:],filename[:]):
    src_data=os.path.join(filepath,datafile)
    with open(src_data,'rb') as file:
         #以binary讀取，資料型態也為byte
        while True:
            e=file.read(13)
            if len(e)<13:
                print('EOF')
                break
            else:
                #item_count-=1
                e=str(e[:4])
                if e not in item_set:
                    item_set.add(e)
                    item_dict[e]=1
                else:
                    item_dict[e]+=1

    item_dict=dict(sorted(item_dict.items(), key=lambda item: item[1],reverse=True))
    df=pd.DataFrame(item_dict.items(),columns=['Element','Count'])
    df.to_csv(os.path.join(filepath,fileindex),index=False)
```

```python
# Generate webdocs_00~19_ground_truth.csv

import sys
sys.path.append("..")

import Tools.Config as TC
import Tools.Func as TF
import Node.DS as DS
import numpy as np
import time
import os
import hyperloglog

import pandas as pd
import re

filepath=r"..\dataset\webdocs"
pattern='out_.*'
r=re.compile(pattern)
filelist=list(filter(r.match,os.listdir(filepath)))    
    # dataset file list

# 產生groudtruth的檔名 webdocs_XX_ground_truth.csv
filename=[]
for item in filelist:
    fid=re.search('\d\d', item).group(0)
    filename.append('webdocs_'+fid+'_ground_truth.csv')

for datafile,fileindex in zip(filelist,filename):
    item_dict=dict()
    # start~end-1
    src_data=os.path.join(filepath,datafile)
    with open(src_data,'r') as file:
        while True:
            e=file.readline().strip('\n')
            if not e:
                print("EOF")
                break
            else:
                if e not in item_dict:
                    item_dict[e]=1
                else:
                    item_dict[e]+=1
    
    # to caida_XX_ground_truth.csv
    item_dict=dict(sorted(item_dict.items(), key=lambda item: item[1],reverse=True))
    df=pd.DataFrame(item_dict.items(),columns=['Element','Count'])
    df.to_csv(os.path.join(filepath,fileindex),index=False)
```
