# Kosarak

```python
import pandas as pd
import os
import matplotlib.pyplot as plt

field=['Precision','Execution','TopARE','TopAAE']
'''
field1='Memory'
field2='Precision'
field3='Execution'
field4='TopARE'
field5='TopAAE'
'''
# =============================read csv files=============================
path=r"result"
dataset='kosarak'
arrival=990002
heavy_ratio=1/1000

# Space saving
ssfile='SS\\'+dataset+'\\HH\\SS_result.csv'
ss=pd.read_csv(os.path.join(path,ssfile))
SSspace_usage=ss['Memory'].tolist()

#CMS
cmfile='CM\\'+dataset+'\\HH\\CM_result.csv'
cm=pd.read_csv(os.path.join(path,cmfile))
CMSspace_usage=cm['Memory'].tolist()


# Elastic
elafile='Ela\\'+dataset+'\\HH\\Ela_result.csv'
ela=pd.read_csv(os.path.join(path,elafile))
Elaspace_usage=ela['Memory'].tolist()


# Heavy Guardian
hgfile='HG\\'+dataset+'\\HH\\HG_result.csv'
hg=pd.read_csv(os.path.join(path,hgfile))
HGspace_usage=hg['Memory'].tolist()

#Cyclic
csfile='CS\\'+dataset+'\\HH\\CS_result.csv'
cs=pd.read_csv(os.path.join(path,csfile))
CSspace_usage=cs['Memory'].tolist()

for findex in field:
    plt.figure()
    plt.xticks([j for j in range(0,1100,100)])
    # Memory usage, 0kb~1MB
    plt.xlabel('Space Usage (KB)')
    if findex=='Precision':
        # initialize Figure
        plt.yticks([j for j in range(0,110,10)])
            # precision: 0~100   
        plt.ylabel('Precision (%), φ=1/5000')
        plt.ylim(0,110)
    elif findex=='Execution':
        # throughput
        min_time=min(min(ss[findex].tolist(),cm[findex].tolist(),
                 ela[findex].tolist(),hg[findex].tolist(),cs[findex].tolist()))
        max_throughput=int(arrival/min_time)
        plt.yticks([j for j in range(0,int(max_throughput*1.1),int(max_throughput*1.1/10))])
        plt.ylabel('Throughtput (item/s)')
        plt.ylim(0,int(max_throughput*1.1))
    elif findex=='TopARE':
        #plt.figure(3)
        max_are=max(max(ss[findex].tolist(),cm[findex].tolist(),
                        ela[findex].tolist(),hg[findex].tolist(),cs[findex].tolist()))
        plt.yticks([j for j in range(0,int(max_are+1)*100,int(max_are+1)*10)])
            # ARE: 0~2
        plt.ylabel('HH ARE (%)')
        plt.ylim(0,int(max_are*110))
    elif findex=='TopAAE':
        #plt.figure(4)
        max_aae=max(max(ss[findex].tolist(),cm[findex].tolist(),ela[findex].tolist(),hg[findex].tolist(),cs[findex].tolist()))
        plt.yticks([j/10 for j in range(0,int(max_aae))])
        plt.yticks([j for j in range(0,int(max_aae+1),int(max_aae/10))]) 
        plt.ylabel('HH AAE')
        plt.ylim(0,int(max_aae*1.1))

    # PLOT
    if findex=='Precision' or findex=='TopARE':
        SS_y=(ss[findex]*100).tolist()
        CM_y=(cm[findex]*100).tolist()
        Ela_y=(ela[findex]*100).tolist()
        HG_y=(hg[findex]*100).tolist()
        CS_y=(cs[findex]*100).tolist()
    elif findex=='Execution':
        # throughput
        SS_y=(arrival/ss[findex]).tolist()
        CM_y=(arrival/cm[findex]).tolist() 
        Ela_y=(arrival/ela[findex]).tolist()
        HG_y=(arrival/hg[findex]).tolist()
        CS_y=(arrival/cs[findex]).tolist()
    elif findex=='TopAAE':
        SS_y=(ss[findex]).tolist()
        CM_y=(cm[findex]).tolist() 
        #Ela_y=(ela[findex]).tolist()
        HG_y=(hg[findex]).tolist()
        CS_y=(cs[findex]).tolist()
        
    SS,=plt.plot(SSspace_usage,SS_y,'-g^',label='SS')   
    CM,=plt.plot(CMSspace_usage,CM_y,'-bo',label='CMS')          
    Ela,=plt.plot(Elaspace_usage,Ela_y,'-cs',label='Ela')     
    HG,=plt.plot(HGspace_usage,HG_y,'-mp',label='HG')
    Cyclic,=plt.plot(CSspace_usage,CS_y,'-r*',label='Cyc')

    # Legend
    plt.legend(handles=[Cyclic,CM,SS,Ela,HG],loc='best')
```
