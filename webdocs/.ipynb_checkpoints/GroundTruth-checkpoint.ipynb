```python
import os
import re
import pandas as pd
path='X:\\NTU\\ML-sketch\\dataset\\Webpage\\webdocs'
pattern='out_.*'
r=re.compile(pattern)
fileli=list(filter(r.match,os.listdir(path)))
    # 把ost_split_xx.txt過濾出來
num_line=0
itemdict={}
for item in fileli:
    with open(os.path.join(path,item),'r') as file:
        while True:
            eli=file.readline().strip('\n')
            if not eli:
                break
            else:
                num_line+=1
                if eli not in itemdict:
                    itemdict[eli]=1
                else:
                    itemdict[eli]+=1
print(num_line)
    # 299887139 items
print(len(itemdict))
    # 5267656 distinct elements
df=pd.DataFrame(list(itemdict.items()),columns=['Element', 'Count'])
df=df.reset_index(drop=True)
df=df.sort_values('Count',ascending=False)
df.head(20)

groundtruth='webdocs_ground_truth.csv'
df.to_csv(groundtruth,index=False)

xdf=pd.read_csv(os.path.join(path,groundtruth))
xdf.head(10)
```
